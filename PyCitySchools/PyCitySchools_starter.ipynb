```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#Total Number of Schools
total_schools = school_data_complete["school_name"].nunique()
#List of Schools
schools = school_data_complete["school_name"].unique()
#Total Number of Students in District
total_students = school_data_complete["student_name"].count()
#Total District Budget
total_budget = sum(school_data_complete.groupby(['school_name'])['budget'].mean())
#Score Averages
math_avg = school_data_complete["math_score"].mean()
reading_avg = school_data_complete["reading_score"].mean()
#Students Passing
math_passing = len(school_data_complete.loc[school_data_complete["math_score"] >= 70])
math_passing_per = (math_passing/total_students)*100
reading_passing = len(school_data_complete.loc[school_data_complete["reading_score"] >= 70])
reading_passing_per = (reading_passing/total_students)*100
passing = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)]['student_name'].count()/total_students
math_passing
```

```python
# all_data = [[total_schools], [total_students], [total_budget], [math_avg], [reading_avg], [math_passing_per], [reading_passing_per], [passing]]
# district_df = pd.DataFrame(all_data, index=["Total Schools", "Total Students", "Total Budget", "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"])

district_df = pd.DataFrame({"Total Schools": [total_schools],
    "Total Students": total_students,
    "Total Budget": total_budget,
    "Average Math Score": math_avg,
    "Average Reading Score": reading_avg,
    "% Passing Math": math_passing_per,
    "% Passing Reading": reading_passing_per,
    "% Overall Passing": passing})
                           
district_df
```

```python
# #school name
# by_school = school_data_complete.set_index('school_name').groupby(['school_name'])
# school_name.head()

# #school types
school_types = school_first.loc[:,'type']

#Total Students
school_students = school_count.loc[:, "Student ID"]

# #Total School Budget
school_budget = school_first.loc[:,'budget']

# #School budget per Student
school_bps = school_budget/school_students

# #Average Math Score
school_math_avg = school_mean.loc[:,'math_score']

# #Average Reading Score
school_reading_avg = school_mean.loc[:,'reading_score']

# # % Passing Math
# school_math_pass = school_data_complete.loc[school_data_complete["math_score"] >= 70]
school_math_pass = school_data_complete.loc[school_data_complete["math_score"] >= 70].groupby('school_name')['Student ID'].count()/school_students

# # % Passing reading
school_reading_pass = school_data_complete.loc[school_data_complete["reading_score"] >= 70].groupby('school_name')['Student ID'].count()/school_students

# # Overall Passing Rate
school_passing = (school_math_pass+school_reading_pass)/2
school_passing
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Set bins for spending
bins = [0, 585, 615, 645, 675]
# create group of categories
spending_ranges = ['< $585', "$585-615", "$615-645", "$645-675"]
```

```python
# putting data into categories
school_summary['Spending Ranges (Per Student)'] = pd.cut(school_summary['Per Student Budget'], bins = bins, labels = spending_ranges)
# group the data set by spending ranges
by_spending_range = school_summary.groupby(['Spending Ranges (Per Student)'])
by_spending_range.head()
```

```python
# calculate score and passing rate by Spending Ranges (Per Student)
mean_math_score = by_spending_range["Average Math Score"].mean()
mean_reading_score = by_spending_range["Average Reading Score"].mean()
mean_percent_math = by_spending_range["% Passing Math"].mean()
mean_percent_reading = by_spending_range["% Passing Reading"].mean()
overall_percent_pass = (mean_percent_math + mean_percent_reading)/2
```

```python
#Create dataframe
school_spending = pd.DataFrame({"Average Math Score":mean_math_score,
                                          "Average Reading Score":mean_reading_score,
                                          "% Passing Math":mean_percent_math,
                                          "% Passing Reading":mean_percent_reading,
                                          "% Overall Passing Rate":overall_percent_pass
                                         })
school_spending
```
